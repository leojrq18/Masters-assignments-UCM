```python
# Descargar los datos con pandas
df_red = pd.read_csv(
    'https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv',
    sep=';'
)
df_white = pd.read_csv(
    'https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-white.csv',
    sep=';'
)
df = pd.concat([df_red, df_white])

df.head()
```

```python
feature_names = [
    'fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides',
    'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol'
]


# separar features y target
y = df.pop('quality').values
X = df.copy().values
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)

print('x_train, y_train shapes:', x_train.shape, y_train.shape)
print('x_test, y_test shapes:', x_test.shape, y_test.shape)
print('Some qualities: ', y_train[:5])
```

```python
## Si quiere, puede normalizar las features
from sklearn.preprocessing import StandardScaler

# Normalizar las features
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

# Mostrar las primeras filas de los datasets normalizados
print("x_train normalizado:")
print(x_train_scaled[:5])

print("x_test normalizado:")
print(x_test_scaled[:5])
```

## Cuestión 1: Cree un modelo secuencial que contenga 4 capas ocultas(hidden layers), con más de 60 neuronas  por capa, sin regularización y obtenga los resultados.

```python
model = tf.keras.models.Sequential()
# Código aquí
model.add(layers.Dense(64, activation='relu', input_shape=(11,)))
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(64, activation='relu'))
# Capa de salida
model.add(layers.Dense(1, activation='linear'))
```

```python
# No modifique el código
model.fit(x_train,
          y_train,
          epochs=200,
          batch_size=32,
          validation_split=0.2,
          verbose=1)
```

```python
# No modifique el código
results = model.evaluate(x_test, y_test, verbose=1)
print('Test Loss: {}'.format(results))
```

## Cuestión 3: Utilice el mismo modelo de la cuestión anterior pero añadiendo un callback de early stopping. No es necesario reducir el test loss.

```python
model = tf.keras.models.Sequential()
# Código aquí
# Agregar capas ocultas con más de 60 neuronas por capa con Dropout y L2 regularization
model.add(layers.Dense(64, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.01), input_shape=(11,)))
model.add(layers.Dropout(0.5))
model.add(layers.Dense(64, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.01)))
model.add(layers.Dropout(0.5))
model.add(layers.Dense(64, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.01)))
model.add(layers.Dropout(0.5))
model.add(layers.Dense(64, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.01)))
model.add(layers.Dropout(0.5))
# Capa de salida
model.add(layers.Dense(1, activation='linear'))
```

```python
# Compilación del modelo
# Código aquí
model.compile(
    optimizer='adam',
    loss='mean_squared_error',
    metrics=['mae']
)
```

```python
## definir el early stopping callback
# Código aquí
early_stopping = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

model.fit(x_train,
          y_train,
          epochs=200,
          batch_size=32,
          validation_split=0.2,
          verbose=1,
          callbacks=[early_stopping]) # Código aquí
```

```python
# No modifique el código
results = model.evaluate(x_test, y_test, verbose=1)
print('Test Loss: {}'.format(results))
```

## Cuestión 4: ¿Podría haberse usado otra función de activación de la neurona de salida? En caso afirmativo especifíquela.

Si. En el contexto de un problema de regresión, como es el caso de predecir la calidad del vino, la elección más común para la función de activación de la neurona de salida es la activación lineal. Esto se debe a que estamos interesados en predecir un valor continuo que puede tomar cualquier valor en el rango de los números reales. La activación lineal no aplica ninguna transformación a la salida, dejándola tal cual para que pueda representar un rango continuo de valores.

Sin embargo, otra función de activación que podría ser utilizada en la neurona de salida es ReLU (Rectified Linear Unit). ReLU transforma la salida de manera que cualquier valor negativo se convierte en cero, mientras que los valores positivos permanecen sin cambios. Esto podría ser útil si queremos asegurar que las predicciones sean siempre no negativas. Aunque en la mayoría de los casos de regresión estándar, la activación lineal es preferida, ReLU podría ser una opción válida dependiendo del dominio específico del problema y de las restricciones que se impongan sobre los valores predichos.

En resumen, la activación lineal es la elección más adecuada y comúnmente utilizada para problemas de regresión debido a su capacidad de manejar un rango continuo de valores. No obstante, en situaciones donde es necesario restringir las predicciones a valores no negativos, la función de activación ReLU podría ser considerada como una alternativa.

## Cuestión 8:  Supongamos que queremos entrenar una red para un problema de clasificación de imágenes con las siguientes clases: {'perro','gato','persona'}. ¿Cuántas neuronas y que función de activación debería tener la capa de salida? ¿Qué función de pérdida (loss function) debería usarse?


Para resolver un problema de clasificación de imágenes con tres clases {'perro', 'gato', 'persona'}, la configuración de la capa de salida de la red neuronal debe ser adecuada para la tarea de clasificación multiclase. En este caso, la capa de salida debe tener tantas neuronas como clases haya en el problema. Por lo tanto, la capa de salida debe tener 3 neuronas, una para cada clase.

La función de activación más adecuada para la capa de salida en un problema de clasificación multiclase es la función softmax. La función softmax convierte las salidas de la red en probabilidades, asegurando que la suma de todas las probabilidades sea igual a 1. Esto es útil porque permite interpretar las salidas de la red como probabilidades de pertenencia a cada una de las clases.

En cuanto a la función de pérdida, la más adecuada para un problema de clasificación multiclase es la función categorical_crossentropy. Esta función de pérdida mide la diferencia entre la distribución de probabilidad predicha por la red y la verdadera distribución de probabilidad de las clases. Minimizar esta pérdida durante el entrenamiento ayuda a que el modelo aprenda a predecir correctamente las clases de las imágenes.

En resumen, para este problema de clasificación de imágenes con tres clases, la capa de salida debe tener 3 neuronas con una función de activación softmax, y la función de pérdida adecuada a utilizar es categorical_crossentropy.
